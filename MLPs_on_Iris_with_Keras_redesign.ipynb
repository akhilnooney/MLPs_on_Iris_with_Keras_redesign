```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.layers.normalization import BatchNormalization
from keras.optimizers import Adam
df=pd.read_csv("/home/akhilnooney/JP_NTBK/DATA/Iris.csv")
```

```python
from keras.utils import np_utils
Label_encoding= LabelEncoder()
Label_encoding.fit(df["Species"])
data_transform= Label_encoding.transform(df["Species"])
One_hot_encoding= np_utils.to_categorical(data_transform)
```

```python
X=df.iloc[:,1:-1].values
Y=One_hot_encoding
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.33, random_state=42)
```

```python
def seq_model(latent=4):
    model = Sequential()
    model.add(Dense(10,activation='relu', input_dim=latent))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(8,activation='relu'))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss = 'categorical_crossentropy', optimizer =Adam(lr = 0.01), metrics = ['accuracy'])
    return model
```

```python
model=seq_model()
history=model.fit(X_train,Y_train, epochs=400, batch_size=70,validation_split=0.10)
```

```python
model.summary()
```

```python
from statistics import mean 
print(f'Train_Accuracy {round((mean(history.history["accuracy"])*100),2)}')
```

```python
_, accuracy = model.evaluate(X_test, Y_test)
```

```python
print(f'Test_Accuracy {round(accuracy*100,1)}')
```
